```python
import pandas,requests,re,time,random
from  bs4 import BeautifulSoup

session=requests.Session()
header={"User-Agent":"Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:59.0) Gecko/20100101 Firefox/59.0",
        "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8"}

ttl_url=[]

for page in range(1,10):
    url = 'http://baa.bitauto.com/baomax3/index-all-all-'+str(page)+'-1.html'
    req = session.get(url, headers=header, timeout=4)

    req_bs4=BeautifulSoup(req.text)

    for content_simple in req_bs4.findAll('div',{'class':'postslist_xh'}):
        temp=list(re.findall(r'href=\"(.*?)\" target=\"_blank\" title=\"(.*?)\"',str(content_simple))[0])
        temp.append(content_simple.find_all('span',{'class':'date'})[0].contents[0].replace(' ','').replace('\r','').replace('\n',''))
        temp.append(content_simple.find_all('a',{'class':'name2'})[0].contents[0].replace(' ','').replace('\r','').replace('\n',''))
        temp.append(page)
        ttl_url.append(temp)
    print(str(url)+'done')
        
ttl_url=pandas.DataFrame(ttl_url)
ttl_url.columns=['url','title','create_date','user','page']
```

```python
import time
date_fix=[]
for date in ttl_url['create_date']:
    if len(re.findall(r'(\d\d:\d\d)',date))>0:
        date_fix.append(time.strftime("%Y-%m-%d"))
    else:
        date_fix.append(date)
ttl_url['create_date']=date_fix
```

## 爬帖子详情

```python
def replace_jkh(text):
    text=text.replace(' ','').replace('\r','').replace('\n','').replace('\xa0','')
    temp=re.findall(r'(<.*?>)',text)
    for i in temp:
        text=text.replace(i,'')
    return text

def content_jx(req,url):
    req_bs4=BeautifulSoup(req.text)
    req_bs4_content=req_bs4.find_all('div',{'class':'post_width'})
    #内容list

    content_list=[]
    for i in req_bs4_content:
        text_test=str(i)
        content_list.append(replace_jkh(text_test))
    #名字list
    req_bs4_content=req_bs4.find_all('a',{'class':'mingzi'})

    name_list=[]
    for i in req_bs4_content:
        text_test=str(i)
        temp=replace_jkh(text_test)
        if len(temp)>0:
            name_list.append(temp)
    #等级list
    req_bs4_content=req_bs4.find_all('a',{'class':'dengji'})

    dengji_list=[]
    for i in req_bs4_content:
        text_test=str(i)
        temp=replace_jkh(text_test)
        if len(temp)>0:
            dengji_list.append(temp)
    min_len=min(len(content_list),len(name_list),len(dengji_list))
    if min_len>0:
        retu_pan=pandas.DataFrame({'content':content_list[:min_len],
                     'name':name_list[:min_len],
                     'dengji':dengji_list[:min_len],})
    else:
        retu_pan=pandas.DataFrame({'content':content_list,
                     'name':name_list,
                     'dengji':dengji_list,})
    retu_pan['url']=url
    retu_pan['ori_url']=url.replace('-1.html','.html')
    return retu_pan
```

```python
def bd_jx(req):
    req_bs4=BeautifulSoup(req.text)
    return_list=[]
    for bsxxx in req_bs4.find_all('div',{'class':'result c-container '}):
        try:
            a=bsxxx
            text_fix=''
            for i in a.find_all('h3',{'class':'t'})[0].find_all('a')[0].contents:
                text_fix=text_fix+str(i).replace('<em>','').replace('</em>','')
            text_fix

            url_temp=re.findall(r'href=\"(.*?)\"',str(a.find_all('h3',{'class':'t'})[0].find_all('a')[0]))[0]

            date_temp=a.find_all('div',{'class':'c-abstract'})[0].find_all('span')[0].contents
            return_list.append([text_fix,url_temp,date_temp])
        except:
            pass
    return return_list
```

## 爬网页内容

```python
def replace_jkh_plus(text):
    text=text.replace(' ','').replace('\r','').replace('\t','').replace('\n','').replace('\xa0','')
    temp=re.findall(r'(<.*?>)',text)
    for i in temp:
        text=text.replace(i,'')
    temp=re.findall(r'[\u4e00-\u9fa5]([^\u4e00-\u9fa5]+?)[\u4e00-\u9fa5]',text)
    for i in temp:
        if len(i)>10:
            text=text.replace(i,'')
    return text
```

```python
from  bs4 import BeautifulSoup
import re

def content_fix(sentence):
    a=list(set(re.findall(r'(<.*?>)',sentence)))
    for b in a:
        sentence=sentence.replace(b,'')
    for b in ['&nbsp']:
        sentence=sentence.replace(b,'')
    return sentence

def jx(sentence):
    a=sentence
    try:
        text=BeautifulSoup(a, "lxml").findAll('div',{'class':'yy_reply_cont'})[0].text
    except:
        try:
            text=BeautifulSoup(a, "lxml").findAll('div',{'class':'tz-paragraph'})[0].text
        except:
            try:
                text=re.findall(r'</div>(.*?)<div',a)[0]
            except:
                text=a
    text=content_fix(text)
    return text
```

```python
split_list=['[','。','，',',','！','……','!','《','》','<','>','\\','"','\'',':','：','？','?','、','|','“','”','‘','’','；',']','{','}','（','）','{','}','【','】','(',')','｛','｝','（','）','：','？','！','。','，',';','、','~','——','+','％','%','`',':','“','”','＂','\'','‘','\n','\r']
```

```python
def split_sentence(sentence):
    temp=[]
    text=''
    for i in sentence:
        if i in split_list:
            if len(text)>0:
                temp.append(text)
            text=''
        else:
            text=text+str(i)
    if len(text)>0:
        temp.append(text)
    return temp
```

##### 提短语

```python
import pandas

emotion_list=ck_qg
feature_list=ck_tz


#二元组提取
result=[]
zx=[]

for m in range(len(word_pan)):
#for m in range(100):
    i=word_pan.iloc[m,0]
    iftz=0
    ifqg=0
    negate = ''
    zx_temp=[]
    zx_temp_cfpd=[]
    for j in i:
        if j in feature_list:
            iftz=iftz+1
            feature_temp=j
            if j not in zx_temp_cfpd:
                zx_temp_cfpd.append(j)
                zx_temp.append([j,''.join(i)])
        if j in emotion_list:
            ifqg=ifqg+1
            emotion_temp=j
        if j in ['不','无','非','没那么','不会','没','不再','不是']:
            negate=j
    if (iftz==1)&(ifqg==1):
        short_sentence_fix=''
        for j in i:
            short_sentence_fix=short_sentence_fix+str(j)
        result.append([feature_temp,negate,emotion_temp,short_sentence_fix,i])
    elif iftz>0:
        zx.extend(zx_temp)
    if m%10000==0:
        print(str(m)+'/'+str(len(word_pan)))

double_result=pandas.DataFrame(result,columns=['特征词','否定词','情感词','短句','短句分词'])
double_result
```
